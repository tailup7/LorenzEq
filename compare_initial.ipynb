```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# change initial condition slightly

# definition of LorenzEq
def lorenz(state, t, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# calculation
def solve_lorenz(initial_conditions, t, sigma, rho, beta):
    return odeint(lorenz, initial_conditions, t, args=(sigma, rho, beta))

# set parameter
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# set initial condition (2 cases)
initial_conditions_1 = [1.00, 1.00, 1.00]  
initial_conditions_2 = [1.00, 1.01, 1.00]  

# set time
t = np.linspace(0, 30, 10000)  # start, end, steps  

# calculation
solution_1 = solve_lorenz(initial_conditions_1, t, sigma, rho, beta)
solution_2 = solve_lorenz(initial_conditions_2, t, sigma, rho, beta)

# set result (time evolution of x)
x1 = solution_1[:, 0]
x2 = solution_2[:, 0]

# output
plt.figure(figsize=(10, 6))
plt.plot(t, x1, label="(x(0),y(0),z(0))=(1.00, 1.00, 1.00)", color='blue')
plt.plot(t, x2, label="(x(0),y(0),z(0))=(1.00, 1.01, 1.00)", color='purple', linestyle='--')
plt.title("Time evolution of LorenzEq's solution is sensitive to initial condition") 
plt.xlabel("Time")
plt.ylabel("x-coordinate")
plt.legend()
plt.grid(True)

# save figure
plt.savefig('compare_initial.png', dpi=300)  
plt.show() 
```

```python
# change constant number in LorenzEq slightly

# set parameter
sigma1 = 10.00
rho1 = 28.0
beta1 = 8.0 / 3.0

sigma2 = 10.01
rho2 = 28.0
beta2 = 8.0 / 3.0

# set initial condition (2 cases)
initial_conditions = [1.00, 1.00, 1.00]  

# set time
t = np.linspace(0, 30, 10000)  # start, end, steps  

# calculation
solution_1 = solve_lorenz(initial_conditions, t, sigma1, rho1, beta1)
solution_2 = solve_lorenz(initial_conditions, t, sigma2, rho2, beta2)

# set result (time evolution of x)
x1 = solution_1[:, 0]
x2 = solution_2[:, 0]

# output
plt.figure(figsize=(10, 6))
plt.plot(t, x1, label="constant number 1 (x)", color='green')
plt.plot(t, x2, label="constant number 2 (x)", color='lime', linestyle='--')
plt.title("Lorenz System: Difference in x-coordinate with slightly different constant number")
plt.xlabel("Time")
plt.ylabel("x-coordinate")
plt.legend()
plt.grid(True)
plt.show()
```
